```python
import numpy as np
from scipy.optimize import minimize

EI = [0,0.02,0,0,0,-0.31,2] #degree 6 polynomial
length = 10
xs = np.linspace(0,length,11)
y_axis_point = [0,1,0]


y_def = [1.6411591496469569, 0.9341104053732115, 0.4164498896636746, 0.13659869228556798, 0.03158185212325293, 0.0057878626629890295, -0.006483606739097647, -0.020076099005160275, -0.020725053247649727, -0.006294831941380474, 1.7541523789077473e-14]

z_def = [-4.37642439905747, -2.4909610809943223, -1.110533039102393, -0.36426317942760456, -0.08421827232825585, -0.015434300434360892, 0.017289617971072424, 0.05353626401378442, 0.05526680866034006, 0.016786218510183915, 5.5067062021407764e-14]

#club_point = [10,0,0]
r_p = [0,0,0]

s = [[0,y_def[0],z_def[0]],[1,y_def[1],z_def[1]],[2,y_def[2],z_def[2]],[3,y_def[3],z_def[3]],[4,y_def[4],z_def[4]],[5,y_def[5],z_def[5]],[6,y_def[6],z_def[6]],[7,y_def[7],z_def[7]],[8,y_def[8],z_def[8]],[9,y_def[9],z_def[9]],[10,y_def[10],z_def[10]]]

w=[]
w_x = []
w_y = []
w_z = []
print(s)
```

```python
for i in range(len(s)):
    ####a = np.array([[x_unit[0],y_unit[0],z_unit[0]], [x_unit[1],y_unit[1],z_unit[1]],[x_unit[2],y_unit[2],z_unit[2]]])
    a = np.array([[1,0,0], [0,1,0],[0,0,1]])
    axis_point = [element * i for element in x_unit]
    b = np.array([s[i][0]-axis_point[0],s[i][1]-axis_point[1],s[i][2]-axis_point[2]])
    x = np.linalg.solve(a, b)
    w.append(x[0])
print(w)
```

```python
for i in range(len(w)):
    w_y.append(w[i][1])
    w_z.append(w[i][2])
```

```python
y = np.polyfit(xs,w_y,8)
z = np.polyfit(xs,w_z,8)
print("y is", y)
```

```python
dy = np.polyder(y)
while len(dy)<8:
    dy = np.insert(dy, 0, 0, axis=0)

d2y = np.polyder(dy)
while len(d2y)<7:
    d2y = np.insert(d2y, 0, 0, axis=0)
    
dz = np.polyder(z)
while len(dz)<8:
    dz = np.insert(dz, 0, 0, axis=0)
d2z = np.polyder(dz)
while len(d2z)<7:
    d2z = np.insert(d2z, 0, 0, axis=0)
print("dy is", dy)
print("d2y is", d2y)
```

```python
M_y = []
M_z = []
for x in xs:
    yresult = d2y[0]*x**6+d2y[1]*x**5+d2y[2]*x**4+d2y[3]*x**3+d2y[4]*x**2+d2y[5]*x+d2y[6]
    zresult = d2z[0]*x**6+d2z[1]*x**5+d2z[2]*x**4+d2z[3]*x**3+d2z[4]*x**2+d2z[5]*x+d2z[6]
    #print("M_y/EI", zresult)
    stiffness = EI[0]*x**6+EI[1]*x**5+EI[2]*x**4+EI[3]*x**3+EI[4]*x**2+EI[5]*x+EI[6]
    #print(stiffness)
    M_y.append(zresult*stiffness)
    M_z.append(yresult*stiffness)

print(M_y)
print(M_z)
```

```python
def f_guess_error(fp):
    fp_y = fp[0]
    fp_z = fp[1]
    full_fp = [0,fp[0],fp[1]]
    
    L_p = []
    for i in range(0,3):
        L_p.append(s[-1][i]-s[0][i])
    L_p = np.linalg.solve(a,L_p)

    
    moment_grip = np.cross(L_p,full_fp)
    moment_head = np.cross(r_p,full_fp)
    print("L_p", L_p)
    print("r_p", r_p)
    #print("full_fp", full_fp)
    print("grip moment", moment_grip)
    print("head moment", moment_head)
    y_error = 0
    z_error = 0
    for i in range(len(xs)):
        #result_x = ((length-xs[i])*moment_grip[0])/length+(xs[i]/length)*moment_head[0]
        result_y = ((length-xs[i])*moment_head[1])/length+(xs[i]/length)*moment_grip[1]
        result_z = ((length-xs[i])*moment_head[2])/length+(xs[i]/length)*moment_grip[2]
        print("m_y", result_y,M_y[i])
        print("m_z", result_z, M_z[i])
        #x_error += abs(result_x - M_x[i])
        y_error += abs(result_y - M_y[i])
        z_error += abs(result_z - M_z[i])
    tot_error = y_error+z_error
    return tot_error
    
    
```
